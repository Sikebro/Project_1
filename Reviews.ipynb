```python
#Importing and applying all dependencies 
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Getting path from the resouce file and reading the information and putting into a dataframe
air_bnb_data = Path("Resources/Listings.csv")
air_bnb_data_df = pd.read_csv(air_bnb_data, encoding = "ISO-8859-1")
```

```python
#Displaying the dataframe to see what information is there to work with.
air_bnb_data_df.head(10)
```

```python
#creating dataframe for number of host for years 2008,2009,2010 & 2019,2020,2021
#For the year 2008
host_2008_start = "2008-01-01"
host_2008_end = "2008-31-12"
host_2008 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2008_start) & (air_bnb_data_df["host_since"] <= host_2008_end))]
count_2008 = host_2008["host_since"].count()
#For the year 2009
host_2009_start = "2009-01-01"
host_2009_end = "2009-31-12"
host_2009 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2009_start) & (air_bnb_data_df["host_since"] <= host_2009_end))]
count_2009 = host_2009["host_since"].count()
#For the year 2010
host_2010_start = "2010-01-01"
host_2010_end = "2010-31-12"
host_2010 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2010_start) & (air_bnb_data_df["host_since"] <= host_2010_end))]
count_2010 = host_2010["host_since"].count()

#For the year 2019
host_2019_start = "2019-01-01"
host_2019_end = "2019-31-12"
host_2019 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2019_start) & (air_bnb_data_df["host_since"] <= host_2019_end))]
count_2019 = host_2019["host_since"].count()
#For the year 2020
host_2020_start = "2020-01-01"
host_2020_end = "2020-31-12"
host_2020 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2020_start) & (air_bnb_data_df["host_since"] <= host_2020_end))]
count_2020 = host_2020["host_since"].count()
#For the year 2021
host_2021_start = "2021-01-01"
host_2021_end = "2021-31-12"
host_2021 = air_bnb_data_df.loc[((air_bnb_data_df["host_since"] >= host_2021_start) & (air_bnb_data_df["host_since"] <= host_2021_end))]
count_2021 = host_2021["host_since"].count()
```

```python
#Average review score for the years of 2008,2009,2010 and 2019,2020,2021

#Average review score for 2008
average_2008 = host_2008["review_scores_rating"].mean()
average_2008_max = host_2008["review_scores_rating"].max()
average_2008_min = host_2008["review_scores_rating"].min()
avg_08 = average_2008.round()
#Average review score for 2009
average_2009 = host_2009["review_scores_rating"].mean()
average_2009_max = host_2009["review_scores_rating"].max()
average_2009_min = host_2009["review_scores_rating"].min()
avg_09 = average_2009.round()
#Average review score for 2010
average_2010 = host_2010["review_scores_rating"].mean()
average_2010_max = host_2010["review_scores_rating"].max()
average_2010_min = host_2010["review_scores_rating"].min()
avg_10 = average_2010.round()
#Average review score for 2019
average_2019 = host_2019["review_scores_rating"].mean()
average_2019_max = host_2019["review_scores_rating"].max()
average_2019_min = host_2019["review_scores_rating"].min()
avg_19 = average_2019.round()
#Average review score for 2020
average_2020 = host_2020["review_scores_rating"].mean()
average_2020_max = host_2020["review_scores_rating"].max()
average_2020_min = host_2020["review_scores_rating"].min()
avg_20 = average_2020.round()
#Average review score for 2021
average_2021 = host_2021["review_scores_rating"].mean()
average_2021_max = host_2021["review_scores_rating"].max()
average_2021_min = host_2021["review_scores_rating"].min()
avg_21 = average_2021.round()
```

```python
#Printing out how host are in each year bracket
print(f"There were {count_2008} hosts who joined in 2008")
print(f"With an average rating of {avg_08} a Maximum of {average_2008_max} and Minimum of {average_2008_min}")
print(f"\nThere were {count_2009} hosts who joined in 2009")
print(f"With an average rating of {avg_09} a Maximum of {average_2009_max} and Minimum of {average_2009_min}")
print(f"\nThere were {count_2010} hosts who joined in 2010")
print(f"With an average rating of {avg_10} a Maximum of {average_2010_max} and Minimum of {average_2010_min}")
print(f"\nThere were {count_2019} hosts who joined in 2019")
print(f"With an average rating of {avg_19} a Maximum of {average_2019_max} and Minimum of {average_2019_min}")
print(f"\nThere were {count_2020} hosts who joined in 2020")
print(f"With an average rating of {avg_20} a Maximum of {average_2020_max} and Minimum of {average_2020_min}")
print(f"\nThere were {count_2021} hosts who joined in 2021")
print(f"With an average rating of {avg_21} a Maximum of {average_2021_max} and Minimum of {average_2021_min}")
```

```python
#Creating a Bar chart to display the number of hosts
x_axis = ["2008","2009","2010","2019","2020","2021"]
y_axis = [count_2008,count_2009,count_2010,count_2019,count_2020,count_2021]

#Creating the bar plot with color, transparency and alignemnt
plt.bar(x_axis,y_axis, color="green", alpha = .75, align = "center")
#Creating the ticks for the bar chart x axis
tick_loc = [value for value in x_axis]
plt.xticks(tick_loc ,x_axis, rotation=60)

plt.title("Number Of Hosts That Joined In Given Year")
plt.xlabel("Year Joined")
plt.ylabel("Number of Hosts")
#plt.savefig("Fig1.png")
plt.show()

```
